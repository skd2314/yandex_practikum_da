### Рассчитаем Retention Rate
<a name="3.4"></a>

Определим время первого визита:

```python
first_activity_date = visits.groupby('uid')['start_ts'].min()
first_activity_date.name = 'first_activity_date'
```

Объединим `visits` и `first_activity_date`:

```python
visits = visits.join(first_activity_date, on='uid', how='left')
visits.head()
```

Определим дату первой активности:

```python
visits['first_activity_date'] = visits['first_activity_date'].astype('datetime64[M]')
visits['first_month'] = visits['session_date'].astype('datetime64[M]')
```

Рассчитаем `lifetime` когорты:

```python
visits['cohort_lifetime'] = visits['first_month'] - visits['first_activity_date']
```

```python
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
```

```python
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype(int)
```

Сгруппируем данные когорты

```python
cohorts = visits.groupby(['first_activity_date', 'cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
```

```python
cohorts.head()
```

Чтобы найти `Retention Rate`, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждую следующую неделю.
Найдём исходное количество пользователей в когорте. Возьмём их число на нулевую неделю:

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_date', 'uid']]
```

Переименуем столбец `uid` в `cohort_users`:

```python
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
```

Объединим данные по когортам с исходным количеством пользователей в когорте:

```python
cohorts = cohorts.merge(initial_users_count, on='first_activity_date')
```

```python
cohorts.head()
```

Рассчитаем `Retention Rate`. Разделим количество активных пользователей в каждую из недель на исходное число пользователей в когорте:

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users'] 
```

Построим сводную таблицу и создадим тепловую карту:

```python
retention_pivot = cohorts.pivot_table(index='first_activity_date', columns='cohort_lifetime',
                                                            values='retention', aggfunc='sum').round(3)
```

```python
retention_pivot.fillna('')
```

```python
sns.set(style='white')
plt.figure(figsize=(16, 10))
plt.title('Retention Rate по окогортам', fontsize=15)
ax=sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', cmap="viridis", vmax=0.09)
ax.set_yticklabels(retention_pivot.index.date)
plt.xlabel('cohort_lifetime', fontsize=15)
plt.ylabel('Когорты', fontsize=15);
```

```python
mean_retention = retention_pivot[1].mean() * 100
print('Средний Retention Rate на второй месяц «жизни» когорт:', mean_retention.round(2), '%')
```

**Вывод:**
* `Retention Rate` весьма низкий на всем периоде, возвращается от 1,9% до 8,5%, т.е. теряем более 90%.

## Маркетинговые метрики
<a name="5"></a>

Получим месяц первой покупки каждого покупателя:

```python
##first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
#first_orders = orders[['uid', 'first_order_month']]

first_orders = orders.groupby('uid').agg({'order_month':'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
first_orders.head()

```

Посчитаем количество новых покупателей `n_buyers` за каждый месяц:

```python
cohort_sizes = (visits_to_orders.groupby('first_order_month').agg({'uid':'nunique'}).reset_index())
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes
```

Построим когорты. Добавим месяц первой покупки каждого покупателя в таблицу с заказами:

```python
orders_new = pd.merge(orders, first_orders, on='uid')
orders_new.head()
```

```python
# удалим дубль first_order_month_y
orders_new.drop(['first_order_month_y'], axis='columns', inplace=True)

orders_new.rename(columns={'first_order_month_x':'first_order_month'}, inplace=True)

orders_new.head()
```

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку

```python
cohorts_ltv = (orders_new.groupby(['first_order_month', 'order_month']).agg({'revenue':'sum'}).reset_index())
cohorts_ltv.columns = ['first_order_month', 'order_month', 'revenue']
cohorts_ltv.head()
```

Добавим в таблицу `cohorts` данные о том, сколько людей первый раз совершили покупку в каждый месяц:

```python
report = pd.merge(cohort_sizes, cohorts_ltv, on='first_order_month')
report.head()
```

Рассчитаем валовую прибыль, умножив выручку на маржинальность.\
Маржинальность сервиса — 100%

```python
margin_rate = 1
report['gp'] = report['revenue'] * margin_rate
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report.head()
```

### Рассчитаем `LTV` по когортам 
Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте:

```python
report['ltv'] = report['gp'] / report['n_buyers']
output = report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean').round(2)
output.cumsum(axis=1).round(2).fillna('')# добавил output.cumsum(axis=1).round(2) для нарастающего итога
```

Отобразим накопительную тепловую карту `LTV` кумулятивно по когортам:

```python
ltv_output = report.pivot_table(index='first_order_month', columns='age', values='ltv').round(2)

sns.set(style='white')
plt.figure(figsize=(16, 10))
plt.title('LTV по когортам', fontsize=15)
ax = sns.heatmap(ltv_output.cumsum(axis=1), annot=True, fmt='.2f', linewidths=1, linecolor='gray', cmap="viridis")
ax.set_yticklabels(ltv_output.index.date)
plt.xlabel('age', fontsize=15)
plt.ylabel('Когорты', fontsize=15);
```

Сложим их `LTV` по месяцам:

```python
ltv_170601 = output.loc['2017-06-01'].sum()

ltv_170601
```
