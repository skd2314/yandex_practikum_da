### Определим платформы с наибольшими суммарными продажами:

Построим график распределения платформ по суммарному количеству проданныъх игр:

```python
plt.figure(figsize=(16, 6))
group_top_sales = df.groupby('platform').sales.sum().sort_values().plot(kind='bar', color='lightblue', ec='black', rot=45)

plt.title('', size=16)
plt.xlabel('Платформа', fontsize=15)
plt.ylabel('Суммарное количество продаж', fontsize=15)
plt.show()
```

У каждой платформы имеется дата начала и окончания жизненого цикла. Вычтем из даты окончания дату начала жизненного цикла. Это необходимо для того чтобы не учитывать далее в анализе "мертвые" платформы

```python
ttl_platform_min = df.pivot_table(index = 'platform', values = 'year_of_release', 
                            aggfunc = ['min']).round(2)
#ttl_platform_min
```

```python
ttl_platform_max = df.pivot_table(index = 'platform', values = 'year_of_release', 
                            aggfunc = ['max']).round(2)
#ttl_platform_max
```

```python
ttl_platform  = ttl_platform_max['max'] - ttl_platform_min['min']
ttl_platform_mean = ttl_platform.sort_values('year_of_release', ascending=False).mean().round(2)

ttl_platform_mean
```

Средняя продолжительность жизненного цикла примерно равна 6,5 лет

Определим % продаж по платформам:

```python
platform_sales_p = (df.groupby("platform").agg({"sales":"sum"})
         .apply(lambda x: 100*x/x.sum()).sort_values("sales", ascending=False)).round(2)
#platform_sales_p.columns = ['% продаж']
platform_sales_p.head(5)
```

Построим таблицу старта продаж, окончания продаж, жизненого цикла и процента от продаж для дальнейшего расчета актуальных top платформ по продажам

```python
ttl_platform_joined = ttl_platform_min.join(ttl_platform_max, how='outer')#min+max
ttl_top_joined2 = ttl_platform_joined.join(ttl_platform, how='outer')#max-min
ttl_top_all = ttl_top_joined2.join(platform_sales_p, how='outer')
ttl_top_all.columns = ['старт_продаж', 'окончание_продаж', 'жизненный_цикл', 'процент_от_продаж']
ttl_top_all.sort_values('процент_от_продаж', ascending=False)
```

```python
top_sales_platform = ttl_top_all.query("жизненный_цикл >= 1 and процент_от_продаж > 3 and окончание_продаж >= 2016")\
        .pivot_table(index='platform').sort_values('жизненный_цикл', ascending=True)
top_sales_platform.sort_values('процент_от_продаж', ascending=False)
```

По результату видим шесть актуальных платформ: X360, PS3, Wii, PS4, 3DS, PC\

## Проверка гипотез 

<a name="6.1"></a>
### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые:

Объявим переменные с необходимыми параметрами:

```python
xbox_q = df.query('platform == "XOne" and user_score > 0')['user_score']
```

```python
pc_q = df.query('platform == "PC" and user_score > 0')['user_score']
```

Установим критический уровень значимости alpha равный 5%, т.е. ошибка не должна превышать 5%, а верный результат не должен быть менее 95%

H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые\
H1: Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
alpha = 0.05

results  = st.ttest_ind(xbox_q, pc_q)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Проверка гипотезы показала, что средние пользовательские рейтинги платформ Xbox One и PC разные

<a name="6.2"></a>
### Средние пользовательские рейтинги жанров Action и Sports разные

Объявим переменные с необходимыми параметрами:

```python
action_q = df.query('genre == "Action" & user_score > 0')['user_score']
action_q.mean()
```

```python
sports_q = df.query('genre == "Sports" & user_score > 0')['user_score']
sports_q.mean()
```

Установим критический уровень значимости alpha равный 5%, т.е. ошибка не должна превышать 5%, а верный результат не должен быть менее 95%

Н0: Средние пользовательские рейтинги жанров Action и Sports одинаковые\
H1: Средние пользовательские рейтинги жанров Action и Sports различаются

```python
alpha = 0.05
results  = st.ttest_ind(action_q, sports_q)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Проверка гипотезы показала, что средние пользовательские рейтинги жанров Action и Sports не различаются
